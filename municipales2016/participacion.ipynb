```python
def get_participacion():
    base_url = 'http://www.servelelecciones.cl/data/participacion/computo/regiones/30'
    urls = []
    for x in range(1, 16):
        if x<10:
            region = '0'+str(x)
        else:
            region = str(x)
        urls.append(base_url+region+'.json')
    
    df = pd.DataFrame()
    for url in urls:
        r = requests.get(url).json()
#         reg_df = pd.DataFrame(r['data'], index=ret_list_cod(r))
        reg_df = pd.DataFrame(r['data'])
        df = pd.concat([df, reg_df])
    return df
        
def ret_list_cod(r):
    comunas = [x['a'] for x in r['data']]
    return [clcomuna.get_steps(x, True) for x in comunas]
    
```
