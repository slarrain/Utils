```python
import pandas as pd
import json
import requests
```

```python
url_json_comunas = "http://www.servelelecciones.cl/data/primaria_alcalde/filters/comunas/all.json"
url_detalle_comuna = "http://www.servelelecciones.cl/data/primaria_alcalde/computo/comunas/"
url_participacion_comuna = "http://www.servelelecciones.cl/data/participacion/computo/comunas/"
```

```python
r = requests.get(url_json_comunas)
```

```python
lista_comunas = r.json()
```

```python
votacion_comunas = {}
lista_doble_primarias = []
for comuna in lista_comunas:
    url = url_detalle_comuna + str(comuna['c'])+".json"
    try:
        req = requests.get(url).json()
    except Exception as e:
        print ("Error on ", comuna, e)
    votacion_comunas[comuna['c']] = req['data']
    if len(req['data'])>1:
        lista_doble_primarias.append(comuna['c']) #Tuvo 2 primarias de la nueva Mayoria y de Chile Vamos
    
    
```

```python
lista_doble_primarias
```

```python
len(votacion_comunas)
```

```python
partipacion = {}
votos = 0
total = 0
for comuna in lista_doble_primarias:
    url = url_participacion_comuna + str(comuna)+".json"
    try:
        req = requests.get(url).json()
    except Exception as e:
        print ("Error on ", comuna, e)
    partipacion[comuna] = req['data']
    for x in req['data']:
        votos += int(x['d'].replace(".", ""))
        total += int(x['c'].replace(".", ""))
```
