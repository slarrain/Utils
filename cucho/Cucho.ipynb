```python
import pandas as pd
import matplotlib as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = (16.0, 12.0)
```

```python
filename = "/home/santiago/Downloads/ClientPC1-PC7.txt"
```

```python
df = pd.read_csv(filename,skiprows=[0,1,2], skipfooter=6, header=None, 
#                  delimiter='  ',
                sep=r'\s+',)
df.rename(columns={4:'Transfer', 6:'Bandwidth', 5:'transfer_metric'}, inplace=True)
df['seconds'] = df[2].str.split('-').apply(lambda x: x[-1]).astype(float)
df.Transfer = df.Transfer.astype(float)
df.Transfer = df.loc[:, ['Transfer', 'transfer_metric']].apply(lambda x: 0.001*x[0] if x[1] == 'KBytes' else x[0], axis=1)
```

```python
plot_title = filename.split('/')[-1].split('.')[0]
```

```python
ax = df.plot.line(x='time', y='Transfer', color='green', title=plot_title)
ax.set_ylabel()
# plt.savefig('')
```
